# Exercício 06 Para cada conjunto de dados fornecido:

(a) Construa uma tabela de distribuição de frequência usando o número de classes indicado.

(b) Inclua colunas registrando os pontos médios das classes, as frequências relativas e acumuladas.

(c) Construa os histogramas de frequência e de frequência relativa.

d) Construa o polígono de frequências e a ogiva.
Nos itens a seguir, estime a proporção dos dados cujos registros:

(e) São maiores ou iguais aqueles associados à classe de maior frequência.

(f) Estão associados à classe de menor frequência.


```python
import pandas as pd
import numpy as np
from math import ceil
import matplotlib.pyplot as plt
#(a)Conjunto de dados 01
vendas = [2114, 2468, 7119, 1876, 4105, 3183,
          1932, 1355, 4278, 1030, 2000, 1077,
          5835, 1512, 1697, 2478, 3981, 1643,
          1858, 1500, 4608, 1000] 
#(a)Conjunto de dados 02
ardencia = [35, 51, 44, 42, 37, 38, 36, 39, 44,
            43, 40, 40, 32, 39, 41, 38, 42, 39,
            40, 46, 37, 35, 41, 39]
#(a)Conjunto de dados 03
reacao = [507, 389, 305, 291, 336, 310, 514, 442,
          373, 428, 387, 454, 323, 441, 388, 426,
          411, 382, 320, 450, 309, 416, 359, 388,
          307, 337, 469, 351, 422, 413]
#(a)Conjunto de dados 04
prova = [1595, 1472, 1820, 1580, 1804, 1635,
         1959, 2020, 1480, 1250, 2083, 1522,
         1306, 1572, 1778, 2296, 1445, 1716,
         1618, 1824]
```

```python
#(b)Criacao da tabela CONJ_01
vendastab = pd.Series(vendas)
n_classes = 6
v_max = vendastab.max()
v_min = vendastab.min()
vendas_freq = vendastab.value_counts(bins=6).sort_index()
nova_vendas_freq = pd.DataFrame(vendas_freq)
nova_vendas_freq = nova_vendas_freq.reset_index()
nova_vendas_freq.columns = ['Classe', 'Frequencia']
amplitude = ceil((v_max-v_min)/n_classes)
intervalos = pd.interval_range(start=v_min, end=v_max+amplitude, freq=amplitude)
nova_vendas_freq['Classe'] = intervalos
pts_medios = [inter.mid for inter in intervalos]
nova_vendas_freq['Pontos medios'] = pts_medios
observacoes = len(vendas)
nova_vendas_freq['Frequencia relativa'] = nova_vendas_freq['Frequencia'] / observacoes
nova_vendas_freq['Frequencia acumulada'] = nova_vendas_freq['Frequencia'].cumsum()
nova_vendas_freq
```

```python
# (c) Criacao do histograma frequencia CONJ_01
hist_vendas = vendastab.hist(bins=[inter.left for inter in intervalos]+
                             [nova_vendas_freq['Classe'][n_classes-1].right])
```

```python
# (c) Criacao do histograma frequencia relativa CONJ_01
hist_vendas = vendastab.hist(bins=[inter.left for inter in intervalos]+
                             [nova_vendas_freq['Classe'][n_classes-1].right],
                             weights=np.ones_like(vendastab)/observacoes)
```

```python
# (d) Criacao do polígono de frequência e ogiva CONJ_01
inferiores = [nova_vendas_freq['Classe'][0].left]
superiores = [int.right for int in nova_vendas_freq['Classe']]
x_vendas = inferiores + superiores
y_vendas = [0] + list(nova_vendas_freq['Frequencia acumulada'])
plt.plot(x_vendas, y_vendas, '-ro')
plt.xticks(x_vendas)
plt.grid()
```
